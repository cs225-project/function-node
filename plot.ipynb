```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#数据读取
df =pd.read_excel('./overhead.xlsx')
# df.head()
df['type']
df_300=df.query('size==300')
df_600=df.query('size==600')
df_900=df.query('size==900')
df_1200=df.query('size==1200')
# df
```

```python
# sns.set_theme(style="darkgrid")
plt.figure(figsize=(14, 2))
plt.subplots_adjust(wspace = 0.25, hspace =0)
plt.subplot(1,4,1)
plot1=sns.lineplot(data=df_300,x='ratio',y='time', hue='type',style='type',markers=True)
plot1.set(xlabel="Read ratio",ylabel="Time (ms)")
plt.title('(a) size=300',y=-0.5)
# plt.xlabel('custom')
# plt.ylabel('custom')
plt.subplot(1,4,2)
plot2=sns.lineplot(df_600,x='ratio',y='time', hue='type',style='type',markers=True)
plot2.set(xlabel="Read ratio",ylabel="")
plt.title('(b) size=600',y=-0.5)

plt.subplot(1,4,3)
plot2=sns.lineplot(df_900,x='ratio',y='time', hue='type',style='type',markers=True)
plot2.set(xlabel="Read ratio",ylabel="")
plt.title('(c) size=900',y=-0.5)


plt.subplot(1,4,4)
plot2=sns.lineplot(df_1200,x='ratio',y='time', hue='type',style='type',markers=True)
plot2.set(xlabel="Read ratio",ylabel="")
plt.title('(d) size=1200',y=-0.5)

plt.savefig("./result.pdf", bbox_inches='tight')
# plt.xlabel('custom')
# plt.ylabel('custom')
```

```python
df =pd.read_excel('./latency.xlsx')
plot1=sns.catplot(df,x='operation',y='latency',kind="bar")
plot1.set(xlabel="operation",ylabel="Latency (ms)")
plt.savefig("./result_latency.pdf", bbox_inches='tight')
```
